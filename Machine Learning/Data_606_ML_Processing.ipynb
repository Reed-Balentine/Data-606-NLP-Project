```python
import os
import numpy as np
import pandas as pd
df = pd.read_pickle("processed_data.pkl")
```

```python
from sklearn.model_selection import (train_test_split, 
                                     GridSearchCV,
                                     RandomizedSearchCV)
from sklearn.feature_extraction.text import (CountVectorizer, 
                                             TfidfVectorizer)
from sklearn.naive_bayes import (MultinomialNB,
                                GaussianNB,
                                ComplementNB,
                                BernoulliNB,
                                CategoricalNB)
from sklearn.pipeline import make_pipeline, Pipeline
from sklearn.ensemble import (HistGradientBoostingClassifier,
                              RandomForestClassifier)

from sklearn.svm import (SVC,
                         NuSVC)
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import (classification_report,
                             confusion_matrix, 
                             precision_score, 
                             recall_score, 
                             f1_score)
from sklearn.base import TransformerMixin
from sklearn.preprocessing import FunctionTransformer, StandardScaler
```

```python
X = df.processed_utterance
y = df.speaker
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1)
```

```python
Tfid_Vector = TfidfVectorizer(analyzer='word', stop_words='english', max_features = 850, ngram_range=[1, 1], 
                       binary=False, norm=None, smooth_idf=True, strip_accents=None,
                       sublinear_tf=True, use_idf=False)
Count_Vector = CountVectorizer(analyzer='word', stop_words='english', max_features = 850, ngram_range=[1, 1], 
                       binary=False, strip_accents=None)
```

```python
pipe_tfid_mNB = Pipeline([
    ('tfid', Tfid_Vector),
    ('MultinomialNB', MultinomialNB())
])

pipe_count_mNB = Pipeline([
    ('count', Count_Vector),
    ('MultinomialNB', MultinomialNB())
])
#--
pipe_tfid_gNB = Pipeline([
    ('tfid', Tfid_Vector),
    ('to_dense', FunctionTransformer(lambda x: x.toarray(), accept_sparse=True)),
    ('GaussianNB', GaussianNB())
])

pipe_count_gNB = Pipeline([
    ('count', Count_Vector),
    ('GaussianNB', GaussianNB())
])
#--
pipe_tfid_compNB = Pipeline([
    ('tfid', Tfid_Vector),
    ('ComplementNB', ComplementNB())
])

pipe_count_gNB = Pipeline([
    ('count', Count_Vector),
    ('ComplementNB', ComplementNB())
])
#--
pipe_tfid_bNB = Pipeline([
    ('tfid', Tfid_Vector),
    ('BernoulliNB', BernoulliNB())
])

pipe_count_bNB = Pipeline([
    ('count', Count_Vector),
    ('BernoulliNB', BernoulliNB())
])
#--
pipe_tfid_catNB = Pipeline([
    ('tfid', Tfid_Vector),
    ('to_dense', FunctionTransformer(lambda x: x.toarray(), accept_sparse=True)),
    ('CategoricalNB', CategoricalNB())
])

pipe_count_catNB = Pipeline([
    ('count', Count_Vector),
    ('CategoricalNB', CategoricalNB())
])
#--
pipe_tfid_hgbc = Pipeline([
    ('tfid', Tfid_Vector),
    ('to_dense', FunctionTransformer(lambda x: x.toarray(), accept_sparse=True)),
    ('HGBC', HistGradientBoostingClassifier())
])

pipe_count_hgbc = Pipeline([
    ('count', Count_Vector),
    ('HGBC', HistGradientBoostingClassifier())
])
#--
pipe_tfid_rf = Pipeline([
    ('tfid', Tfid_Vector),
    ('rf', RandomForestClassifier())
])

pipe_count_rf = Pipeline([
    ('count', Count_Vector),
    ('rf', RandomForestClassifier())
])
#----
pipe_tfid_svc = Pipeline([
    ('tfid', Tfid_Vector),
    ('Scaler', StandardScaler(with_mean=False)),
    ('SVC', SVC())
])

pipe_count_svc = Pipeline([
    
    ('count', Count_Vector),
    ('SVC', SVC())
])

pipe_tfid_nusvc = Pipeline([
    ('tfid', Tfid_Vector),
    ('NuSVC', NuSVC())
])

pipe_count_nsvc = Pipeline([
    ('count', Count_Vector),
    ('NuSVC', NuSVC())
])

pipe_tfid_knn = Pipeline([
    ('tfid', Tfid_Vector),
    ('knn', KNeighborsClassifier())
])

pipe_count_knn = Pipeline([
    ('count', Count_Vector),
    ('knn', KNeighborsClassifier())
])

pipe_tfid_dt = Pipeline([
    ('tfid', Tfid_Vector),
    ('dt', DecisionTreeClassifier())
])

pipe_count_dt = Pipeline([
    ('count', Count_Vector),
    ('dt', DecisionTreeClassifier())
])
```
